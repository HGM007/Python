## matplotlib scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([5,2,6,7,8,5,9,3,6])
y=np.array([7,4,3,5,7,7,8,5,4])
plt.scatter(x,y,color='r')
plt.show()
```

```python
#D= 1 age and speed of 9 cars
x=np.array([5,2,6,7,8,5,9,3,6])
y=np.array([90,55,63,55,111,23,78,99,69])
plt.scatter(x,y,color='r')

#D=2  age and speed of 10 cars
x=np.array([4,6,2,7,9,6,5,8,9,4])
y=np.array([45,66,77,33,55,99,77,66,55,121])
plt.scatter(x,y,color='yellow')

plt.show()
```

```python
#x=np.array(np.random.randint(0,range,size))
x=np.array([4,6,2,7,9])
y=np.array([45,66,77,33,55])
color=np.array(["red","green","violet","yellow","pink",])
plt.scatter(x,y,c=color)
plt.show()
```

```python
x=np.array([4,6,2,7,9,7])
y=np.array([45,66,77,33,55,36])
size=np.array([200,310,40,100,60,500])
color=np.array(["red","green","violet","yellow","pink","maroon"])
plt.scatter(x,y,c=color,s=size,alpha=0.3)
plt.show()
#for overlapping use alpha
```

```python
x=np.random.randint(100,size=(100))
y=np.random.randint(100,size=(100))
color=np.random.randint(100,size=(100))
size=10*np.random.randint(100,size=(100))

plt.scatter(x,y,c=color,s=size,alpha=0.4,cmap='nipy_spectral')
plt.colorbar()
plt.show()
```

## Bar plot

```python
x=np.array(["A","B","C","D"])
y=np.array([3,4,6,10])
#color=np.random.randint(3,size=(4))
c=np.array(["red","pink","black","maroon"])
plt.bar(x,y,width=0.5,color=c)
plt.show()
```

## pie plot

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,15,35,15])
mylabel=["hgm","aa","hp","srk"]
# for cut out anf rep use explode
myexp=[0,0.2,0,0.2]
#indication of data color etc
plt.pie(y,labels=mylabel,explode=myexp,shadow=True)
plt.legend(loc="upper left",title="scr")
plt.show()
```
